### 6- Scipy Manipulation

> shapiro wilk test

```python
from scipy.stats import shapiro
data = [0.23, 0.43, 0.53, 0.64, 0.88]
stat,p = shapiro(data)
print ('stat =', stat)
print ('p = ', p)
if p > 0.05:
 print("Data is normal")
else:
 print ("Data is not normal")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.stats import shapiro
ks1 = sns.load_dataset('titanic')
stat,p = shapiro(ks1['age'])
print ('stat =', stat)
print ('p = ', p)
if p > 0.05:
 print("Data is normal")
else:
 print ("Data is not normal")
sns.boxplot(ks1['age'])
#ks1['age'].hist()
```

```python
# Basic fun of open cv

import cv2 as cv
import numpy as np
img = cv.imread('resources/bike.jpg')

# Resize
resize_img = cv.resize(img, (450, 250)) # where 1st is width and 2nd is height

# convert to Gray image
gray_img = cv.cvtColor(img, cv.COLOR_BGR2GRAY) #color conversion

# Blured imag
blur_img = cv.GaussianBlur(img, (19,19), 0)

# Edge detection
edge_img = cv.Canny(img, 97,97)

# Thikness of edge lines
mat_kernal = np.ones((3,3), np.uint8)
dilate_img = cv.dilate(edge_img, (mat_kernal), iterations=1)

# Make thiner edge outlines
ero_img = cv.erode(dilate_img, (mat_kernal), iterations=1)

# Black & whit image
(thresh, binary) = cv.threshold(gray_img, 127, 255, cv.THRESH_BINARY)

# Cro Image numpy lib will be use
print('size of the image : ', img.shape)
crop_img = img[0:400, 0:600 ]

cv.imshow('Orignal', img)
cv.imshow('Resize image', resize_img)
cv.imshow('Gray image', gray_img)
cv.imshow('Blured image', blur_img)
cv.imshow('Edge image', edge_img)
cv.imshow('Thik outlined image', dilate_img)
cv.imshow('Thin outlined image', ero_img)
cv.imshow('Black & white image', binary)
cv.imshow('Cropped image', crop_img)


cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# change the prespective of pic
import cv2 as cv
import numpy as np

img = cv.imread('resources/pic.jpeg')
# Resize
img = cv.resize(img, (450, 550))
#img.shape
width = 550
height = 450

# points
point1 = np.float32([[81, 63], [393, 64], [0,384], [448,380]])

point2 =np.float32([[0,0], [width,0],[0, height], [width, height]])


matrix = cv.getPerspectiveTransform(point1, point2)

out_img = cv.warpPerspective(img, matrix, (width, height))

cv.imshow('Orignal Picture', img)
cv.imshow('Transformed picture', out_img)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# How to find coordinates and color pallate of a image 

import cv2 as cv
import numpy as np

def find_coord(event, x,y, flags, parameter):
    if event == cv.EVENT_FLAG_LBUTTON:
        # Mapping on left mouse click to find coordinates
        print(x, '', y)

        # Text perameters
        font = cv.FONT_HERSHEY_SIMPLEX
        cv.putText(img, str(x) + ',' + str(y),(x, y), font, 1, (233,0,0), thickness=2)


        # show text on image
        cv.imshow('image', img)

    # for color display mapping right mouse click
    if event == cv.EVENT_RBUTTONDOWN:
        print(x, '', y)
        font = cv.FONT_HERSHEY_SIMPLEX
        b = img[y, x, 0]
        g = img[y, x, 1]
        r = img[y, x, 2]

        cv.putText(img, str(b) + ',' + str(g) + ',' + str(r) ,(x, y), font, 1, (32,28,27), thickness=2)
        # show text on image
        cv.imshow('image', img)

# funt to read and display

if __name__=="__main__":
    #reading an image
    img = cv.imread("resources/pic.jpeg", 1)
    img = cv.resize(img, (450, 550)) # resize image
    #Display the image
    cv.imshow('image', img)
    #Setting call back fun on mouse click
    cv.setMouseCallback("image", find_coord)

    cv.waitKey(0)
    cv.destroyAllWindows()
```

```python
# Detect specific color

import cv2 as cv
import numpy as np

path = "resources/bike.jpg"

# Sliders for HSV
def slider():
    pass

cv.namedWindow('Bars')
cv.resizeWindow('Bars', 1000, 600)

cv.createTrackbar('Hue Min', 'Bars', 0,179, slider)
cv.createTrackbar('Hue Max', 'Bars', 179,179, slider)

cv.createTrackbar('Sat Min', 'Bars', 0,255, slider)
cv.createTrackbar('Sat Max', 'Bars', 255,255, slider)

cv.createTrackbar('Value Min', 'Bars', 0,255, slider)
cv.createTrackbar('Value Max', 'Bars', 255,255, slider)

while True:
    img = cv.imread(path)
    hsv_img = cv.cvtColor(img, cv.COLOR_BGR2HSV)
    hue_min = cv.getTrackbarPos('Hue Min', 'Bars')
    hue_max = cv.getTrackbarPos('Hue Max', 'Bars')
    sat_min = cv.getTrackbarPos('Sat Min', 'Bars')
    sat_max = cv.getTrackbarPos('Sat Max', 'Bars')
    val_min = cv.getTrackbarPos('Value Min', 'Bars')
    val_max = cv.getTrackbarPos('Value Max', 'Bars')

    print(hue_min, hue_max, sat_min, sat_max, val_min, val_max)
    
    # Code to see changes in pictures
    lower = np.array([hue_min, sat_min, val_min])
    upper = np.array([hue_max, sat_max, val_max])
    mask_image = cv.inRange(hsv_img, lower, upper)
    out_img = cv.bitwise_and(img, img, mask=mask_image)

    #cv.imshow('Orignal Picture', img)
    #cv.imshow('HSV converted Picture', hsv_img)
    cv.imshow("Mask", mask_image)
    cv.imshow("Final output", out_img)

    if cv.waitKey(1) & 0xFF == ord('q'):
            break
    
cv.destroyAllWindows()
```

```python
# Face detection in Images
import cv2 as cv
face_cascade = cv.CascadeClassifier('resources/haarcascade_frontalface_default.xml')

path = "resources/people_faces.jpg"
img = cv.imread(path)
img = cv.resize(img, (1000, 600))

faces = face_cascade.detectMultiScale(img, 1.1, 4)

# Draw ractangle
for (x,y,w,h) in faces:
    cv.rectangle(img, (x,y), (x+w, y+h), (255,0,0), 2)


cv.imshow('Orignal Image', img)
cv.imwrite("resources/Detected Faces.jpg",img)
cv.waitKey(0)
cv.destroyAllWindows()
```
